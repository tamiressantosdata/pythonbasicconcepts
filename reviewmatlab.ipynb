```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(6)
```

```python
x
```

```python
y=x+1
```

```python
plt.plot(x, y, 'o--')
plt.show()
```

```python
plt.plot(x, y, 'o--')
plt.plot(x, -y, 'o-')
plt.show()
```

```python
#title
plt.plot(x, y, 'o--')
plt.plot(x, -y, 'o-')
plt.title('y=x and y=-x')
plt.show()
```

```python
N = 16
x = np.linspace(0, 15, N)
print(x)
```

```python
y = x
plt.plot(x, y, 'o--')
plt.axis('off')
plt.show()
```

```python
y=np.logspace(0.1, 2, N)
print(y)
plt.plot(x, y, 'o--')
plt.show()
```
